```python
import json
import pandas as pd
import random
import string

# Load SQuAD 2.0 dataset
with open("train-v2.0.json", "r") as f:
    squad_data = json.load(f)

# Extract question-answer pairs
contextual_data = []
for topic in squad_data["data"]:
    for paragraph in topic["paragraphs"]:
        for qa in paragraph["qas"]:
            question = qa["question"]
            answer = qa["answers"][0]["text"] if qa["answers"] else ""
            contextual_data.append({"question": question, "answer": answer, "label": 1})

# Randomly sample 5,000 examples
contextual_data = random.sample(contextual_data, 5000)

# Convert to DataFrame
contextual_df = pd.DataFrame(contextual_data)

# List of random sentences for non-contextual answers
random_sentences = [
    "The sky is blue.",
    "I like pizza.",
    "My favorite color is red.",
    "The cat is on the roof.",
    "I don’t know.",
    "Bananas are yellow.",
    "The sun rises in the east.",
    "I enjoy reading books.",
    "The dog barked loudly.",
    "The weather is nice today."
]

# Create non-contextual examples
non_contextual_data = []
for _, row in contextual_df.iterrows():
    question = row["question"]
    answer = random.choice(random_sentences)
    non_contextual_data.append({"question": question, "answer": answer, "label": 0})

# Convert to DataFrame
non_contextual_df = pd.DataFrame(non_contextual_data)

# Function to generate random strings
def generate_random_string(length=10):
    return ''.join(random.choice(string.ascii_lowercase) for _ in range(length))

# Create nonsensical examples
nonsensical_data = []
for _, row in contextual_df.iterrows():
    question = row["question"]
    answer = generate_random_string()
    nonsensical_data.append({"question": question, "answer": answer, "label": 0})

# Convert to DataFrame
nonsensical_df = pd.DataFrame(nonsensical_data)

# Combine all DataFrames
dataset = pd.concat([contextual_df, non_contextual_df, nonsensical_df], ignore_index=True)

# Shuffle the dataset
dataset = dataset.sample(frac=1).reset_index(drop=True)

# Save the dataset to a CSV file
dataset.to_csv("balanced_question_answer_dataset.csv", index=False)

print("Dataset created successfully!")
```

```python
# Install required libraries
!pip install transformers datasets

# Load and clean the dataset
import pandas as pd

dataset = pd.read_csv("balanced_question_answer_dataset.csv")
dataset["question"] = dataset["question"].fillna("").astype(str)
dataset["answer"] = dataset["answer"].fillna("").astype(str)

# Convert to Hugging Face Dataset
from datasets import Dataset

hf_dataset = Dataset.from_pandas(dataset)
train_test_split = hf_dataset.train_test_split(test_size=0.1)
train_dataset = train_test_split["train"]
val_dataset = train_test_split["test"]

# Tokenize the dataset
from transformers import AutoTokenizer

model_name = "distilbert-base-uncased"  # You can also use "roberta-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tokenize_function(examples):
    return tokenizer(examples["question"], examples["answer"], padding="max_length", truncation=True)

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)

train_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
val_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])

# Load the pretrained model
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

# Set up training arguments
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    save_total_limit=2,
    logging_dir="./logs",
    logging_steps=10,
)

# Define the Trainer
from transformers import Trainer

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# Train the model
trainer.train()

# Evaluate the model
results = trainer.evaluate()
print(results)

# Save the fine-tuned model
model.save_pretrained("./fine-tuned-model")
tokenizer.save_pretrained("./fine-tuned-model")
```

```python
# Load the fine-tuned model and tokenizer
from transformers import pipeline

classifier = pipeline("text-classification", model="/content/drive/MyDrive/fine-tuned-model")

# Test cases
test_cases = [
    {"question": "Q16A. What is the most important thing you LIKE about the shown concept}?     This can include anything you would want kept for sure or aspects that might drive you to buy or try it…       Please type a detailed response in the space below?", "answer": "Calidad y funcionalidad"},  # Contextual
    {"question": "What is the capital of France?", "answer": "Bananas are yellow."},  # Non-contextual
    {"question": "Who wrote 'To Kill a Mockingbird'?", "answer": "Harper Lee"},  # Contextual
    {"question": "Who wrote 'To Kill a Mockingbird'?", "answer": "I like school."},  # Non-contextual
    {"question": "What is 2 + 2?", "answer": "4"},  # Contextual
    {"question": "What is 2 + 2?", "answer": "asdfghjkl"},  # Nonsensical
]

# Run inference on test cases
for test_case in test_cases:
    question = test_case["question"]
    answer = test_case["answer"]
    result = classifier(f"{question} [SEP] {answer}")

    # Extract prediction and score
    label = result[0]["label"]
    score = result[0]["score"]

    # Compute sensical score
    if label == "LABEL_1":
        sensical_score = score  # Sensical answer
    else:
        sensical_score = 1 - score  # Nonsensical answer

    # Display results
    print(f"Question: {question}")
    print(f"Answer: {answer}")
    print(f"Sensical Score: {sensical_score:.2f}")
    print()

# Save results to a CSV file
import pandas as pd

results = []
for test_case in test_cases:
    question = test_case["question"]
    answer = test_case["answer"]
    result = classifier(f"{question} [SEP] {answer}")

    label = result[0]["label"]
    score = result[0]["score"]
    sensical_score = score if label == "LABEL_1" else 1 - score

    results.append({
        "question": question,
        "answer": answer,
        "sensical_score": sensical_score
    })

# Convert to DataFrame and save to CSV
results_df = pd.DataFrame(results)
results_df.to_csv("sensical_scores.csv", index=False)

# Display the saved results
print(results_df)
```
