```python
sample_filepath = "/content/drive/MyDrive/sampleData.xlsx"
```

```python
df0 = pd.read_excel(sample_filepath, sheet_name=0, header=[0, 1], engine="openpyxl")
df0.columns = ['_'.join(map(str, col)).strip() for col in df0.columns]
df0.columns = [col[:col.find('_Unnamed')] if '_Unnamed' in col else col for col in df0.columns]
df0.head()
```

```python
df0.fillna('Unanswered', inplace=True)
if(df0.duplicated(subset=list(df0.columns[2:])).sum() > 0) :
  df0 = df0.drop_duplicates(subset=list(df0.columns[2:]), keep='last')
```

```python
df0.to_csv("sampleData_1.csv", index=False)
```

```python
df = pd.read_excel(sample_filepath, sheet_name=1, header=[0, 1], engine="openpyxl")
df.columns = ['_'.join(map(str, col)).strip() for col in df.columns]
df.columns = [col[:col.find('_Unnamed')] if '_Unnamed' in col else col for col in df.columns]
df.head()
```

```python
duration = ((df['End Date'] - df['Start Date']).dt.total_seconds() / 60).round()
df = df.drop(columns=['Start Date', 'End Date'], axis=1)
df.insert(1, 'Response Duration', duration)
df.head()
```

```python
df.fillna('Unanswered', inplace=True)
df.head()
```

```python
if(df.duplicated(subset=list(df.columns[2:])).sum() > 0) :
  df = df.drop_duplicates(subset=list(df.columns[2:]), keep='last')
```

```python
df.to_csv("sampleData_1.csv", index=False)
```
